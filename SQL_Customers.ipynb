click-through rates for individual people for each category of webpage by selecting data and grouping it appropriately.

```python
%%sql 
SELECT MIN(customers.userid), name, webpageviewed, 
    ROUND(SUM(clicked)*1.0/COUNT(clicked),3) AS click_rate
FROM customers NATURAL JOIN adclicks
WHERE customers.userid = 64 OR customers.userid = 48
GROUP BY name, webpageviewed;
```

It's unlikely that two customers will have the same "first view" time, but joining on the pair will ensure uniqueness in this case.

```python
%%sql
WITH first_view_table AS 
   (SELECT userid, MIN(time) AS first_view 
   FROM customers JOIN adclicks USING(userid)
   GROUP BY userid)
SELECT *
FROM first_view_table JOIN adclicks ON first_view = time AND first_view_table.userid = adclicks.userid
LIMIT 5;
```

count the number of "first views" for each type of website, or compute click-through rates for each type of website, etc.

```python
%%sql

WITH first_view_table AS 
   (SELECT userid, MIN(time) AS first_view 
   FROM customers JOIN adclicks USING(userid)
   GROUP BY userid)
    
SELECT webpageviewed, COUNT(*) AS count_first_views
FROM first_view_table JOIN adclicks ON first_view = time AND first_view_table.userid = adclicks.userid
GROUP BY webpageviewed;
```

Computing click-through rates by website type

```python
%%sql

WITH first_view_table AS 
   (SELECT userid, MIN(time) AS first_view 
   FROM customers JOIN adclicks USING(userid)
   GROUP BY userid)
    
SELECT webpageviewed, SUM(clicked) AS clicks, COUNT(*) AS views, 
   ROUND(AVG(clicked),3) AS first_click_through_rate
FROM first_view_table JOIN adclicks ON first_view = time AND first_view_table.userid = adclicks.userid
GROUP BY webpageviewed;
```

Counting click-through rates by adid

```python
%%sql

WITH first_view_table AS 
   (SELECT userid, MIN(time) AS first_view 
   FROM customers JOIN adclicks USING(userid)
   GROUP BY userid)
SELECT adid, SUM(clicked) AS clicks, COUNT(*) AS views, 
   ROUND(AVG(clicked),3) AS first_click_through_rate
FROM first_view_table JOIN adclicks ON first_view = time AND first_view_table.userid = adclicks.userid
GROUP BY adid;
```

we are generating a categorical variable based on a customer's age. We generate a new age_bracket column in the resulting table by checking the value of age and returning a new value.

```python
%%sql
SELECT userid AS uid, age,
    CASE
        WHEN  0 <= age AND age < 13 THEN 'young'
        WHEN 13 <= age AND age <= 19 THEN 'teens'
        WHEN 20 <= age AND age < 30 THEN '20s'
        WHEN 30 <= age AND age < 40 THEN '30s'
        WHEN 40 <= age AND age < 50 THEN '40s'
        WHEN 50 <= age AND age < 60 THEN '50s'
        WHEN 60 <= age AND age < 70 THEN '60s'
        WHEN 70 <= age THEN 'seniors'
        ELSE 'other'
    END AS age_bracket 
FROM customers
LIMIT 10;
```

```python
%%sql 
WITH brackets AS 
    (SELECT userid AS uid, 
    CASE
        WHEN  0 <= age AND age < 13 THEN 'young'
        WHEN 13 <= age AND age <= 19 THEN 'teens'
        WHEN 20 <= age AND age < 30 THEN '20s'
        WHEN 30 <= age AND age < 40 THEN '30s'
        WHEN 40 <= age AND age < 50 THEN '40s'
        WHEN 50 <= age AND age < 60 THEN '50s'
        WHEN 60 <= age AND age < 70 THEN '60s'
        WHEN 70 <= age THEN 'seniors'
        ELSE 'other'
    END AS age_bracket 
    FROM customers)
    
SELECT age_bracket, 
    ROUND(AVG(clicked),3) AS click_rate, 
    COUNT(clicked) AS number_of_ad_views
FROM brackets, adclicks 
WHERE adclicks.userid = uid
GROUP BY age_bracket 
ORDER BY age_bracket;
```

find click-through rates for different age group/adid combinations.

```python
%%sql 
WITH brackets AS 
    (SELECT userid AS uid, 
    CASE
        WHEN  0 <= age AND age < 13 THEN 'young'
        WHEN 13 <= age AND age <= 19 THEN 'teens'
        WHEN 20 <= age AND age < 30 THEN '20s'
        WHEN 30 <= age AND age < 40 THEN '30s'
        WHEN 40 <= age AND age < 50 THEN '40s'
        WHEN 50 <= age AND age < 60 THEN '50s'
        WHEN 60 <= age AND age < 70 THEN '60s'
        WHEN 70 <= age THEN 'seniors'
        ELSE 'other'
    END AS age_bracket 
    FROM customers)
SELECT adid, age_bracket, 
    ROUND(AVG(clicked),3) AS click_rate, 
    COUNT(clicked) AS number_of_ad_views
    
FROM brackets, adclicks 
WHERE adclicks.userid = uid
GROUP BY adid, age_bracket 
ORDER BY adid, age_bracket
LIMIT 10;
```

```python
%%sql 
SELECT age/10 AS decades_old, 
    ROUND(AVG(clicked),3) AS click_rate, 
    COUNT(clicked) AS number_of_ad_views
    
FROM customers, adclicks 
WHERE customers.userid = adclicks.userid
GROUP BY decades_old 
ORDER BY decades_old;
```

SQL Alchemy

```python
import sqlalchemy
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String

# This creates a base we can inherit from
Base = declarative_base()

class Customers(Base):
    __tablename__ = "customers"
    
    # sqlalchemy uses these names as the column names
    userID = Column(Integer, primary_key=True)
    name = Column(String)
    zipcode = Column(String)
    age = Column(Integer)
```

```python
engine = sqlalchemy.create_engine('sqlite:///small_data/customers.sqlite', echo=True)

session_gen = sqlalchemy.orm.sessionmaker(bind=engine)
```

```python
session = session_gen()
```

```python
%%sql

SELECT userid, name
FROM customers
WHERE userid < 40
ORDER BY userid DESC
LIMIT 7
```

```python
results = session.query(Customers.userID, Customers.name).filter(Customers.userID < 40)\
                  .order_by(sqlalchemy.desc(Customers.userID))[2:7]

for row in results:
    print(row)
```

```python
session.query(Customers.name, Customers.zipcode, Customers.age)\
       .filter(sqlalchemy.or_(Customers.zipcode.like('06%'), Customers.zipcode == None))\
       .filter(Customers.age.in_([19, 20, 21, 22, 23])).all()
```

```python
from sqlalchemy.sql import func

underage = session.query(Customers).filter(Customers.age < 21).count()
print("There are {} people who are under 21".format(underage))

session.query(Customers.zipcode, func.count('*')).filter(Customers.zipcode.like('06%'))\
       .group_by(Customers.zipcode).all()
```

```python
case_part = sqlalchemy.case([(Customers.age < 20, "young"), \
                               (sqlalchemy.and_(Customers.age >= 20, Customers.age < 30), "20s")], \
                            else_="older").label("age_group")

session.query(Customers.name, Customers.age, case_part)[:10]
```
